```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
data = 'https://raw.githubusercontent.com/aritsarann/data-python-nd/main/dads/data/world_air_quality.csv'
air = pd.read_csv(data,sep=';')
df = pd.DataFrame(air)
```

```python
df.head()
```

```python
df = df[df['Pollutant'] == 'PM2.5']

df[['Latitude', 'Longitude']] = df['Coordinates'].str.split(', ', expand=True)
df['Latitude'] = df['Latitude'].astype(float)
df['Longitude'] = df['Longitude'].astype(float)

df = df.dropna(subset=['Latitude', 'Longitude'])
df.drop(columns=["Coordinates",'Location'], inplace=True)
```

```python
df.isna().sum()
```

```python
df.loc[air['Country Label'].isnull() == True]
```

```python
country_labels = {
    'XK': 'Kosovo',
    'KV': 'Kosovo',
    'AJ': 'Azerbaijan',
    'CE': 'Sri Lanka',
    'UC': 'Curacao',
    'CW': 'Curacao',
    'TI': 'Tajikistan',
    'IZ': 'Iraq',
    'KU': 'Kuwait',
    'BK': 'Bosnia and Herzegovina',
    'VM': 'Viet Nam',
    'AQ': 'American Samoa',
    'TX': 'Turkmenistan',
    'MK': 'Macedonia'
}

for code, label in country_labels.items():
    df.loc[df['Country Code'] == code, 'Country Label'] = label

df.isna().sum()
```

```python
# Converts the 'Last Updated' column of a DataFrame to datetime format
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce', utc=True)
df['Year'] = df['Last Updated'].dt.year
df.head(5)
```

```python
df[df.Value < 0].head()
```

```python
df.drop(df.loc[df.Value < 0].index, inplace=True)
df.drop(df.loc[df.Value == 0].index, inplace=True)
```

```python
df2023 = df[(df['Year'] == 2023) | (df['Year'] == 2024)]
df2023.head()
```

```python
df2023['avg_value'] = df2023.groupby(['Year','Latitude', 'Longitude'])['Value'].transform('mean')   #คำนวณค่าเฉลี่ยของ 'Value' ตามกลุ่ม 'Year', 'Latitude', 'Longitude'
df2023.drop_duplicates(subset=['Year','Latitude','Longitude'], inplace=True)
df2023.loc[:, 'Value'] = df2023['avg_value']
df2023.drop(columns = ['avg_value'], inplace=True)

df2023.reset_index(drop=True, inplace=True)
```

```python
df2023.head(5)
```

```python
df2023['Pollution Level'] = pd.cut(df2023['Value'], bins=[0, 12.1, 35.5, 55.5, 150.5, float('inf')],
                                 labels=['Low', 'Moderate', 'Unhealthy for Sensitive Groups','Unhealthy', 'Very Unhealthy'])
df2023['Pollution Level'].unique()
```

```python
# Create Scatter Mapbox
fig = px.scatter_mapbox(df2023,
                        lat='Latitude',
                        lon='Longitude',
                        hover_name='Country Label',
                        hover_data={'Value': True,'Year':True},
                        color='Pollution Level',
                        color_discrete_map={'Low': 'green',
                                            'Moderate': 'yellow',
                                            'Unhealthy for Sensitive Groups': 'orange',
                                            'Unhealthy': 'firebrick',
                                            'Very Unhealthy': 'indigo'},
                        zoom=0.4,
                        center={"lat": 20, "lon": 40},
                        mapbox_style="carto-positron"
                       )

# Show additional information when hovering over points on the map
fig.update_layout(hoverlabel=dict(bgcolor="white", font_size=12))
fig.update_layout(title="World Air Quality 2023-2024 : PM2.5")

# Show the map
fig.show()
```
