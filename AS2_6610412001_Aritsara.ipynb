## 1. Which football teams have the  fewest total number of cards (yellow + red)?

```python
df_y = df[['HomeTeam','AwayTeam','HY','AY','HR','AR']]
```

```python
home_c =  df_y.groupby('HomeTeam')[['HY','HR']].sum().reset_index()
away_c =  df_y.groupby('AwayTeam')[['AY','AR']].sum().reset_index()

# Merge data
c_count = home_c.merge(away_c, how='outer', left_on='HomeTeam', right_on='AwayTeam')
c_count = c_count.drop(['AwayTeam'], axis=1)
c_count.rename(columns={"HomeTeam": "Team"}, inplace=True)
c_count = c_count.set_index('Team')

# Calculate total cards
c_count["Total Yellow Cards"] = c_count["HY"] + c_count["AY"]
c_count["Total Red Cards"] = c_count["HR"] + c_count["AR"]
c_count["Total Cards"] = c_count["Total Yellow Cards"] + c_count["Total Red Cards"]

# Display the team that received the fewest total cards
c_count[c_count['Total Cards'] == c_count['Total Cards'].min()].iloc[:,-3:]
```

## 2. Which team has the highest average number of goals scored per game this season? 
average number of goals scored per game

```python
df1 = df[['HomeTeam','AwayTeam','FTHG','FTAG']]
```

```python
home_goals = df1.groupby('HomeTeam')['FTHG'].sum().reset_index(name='Home Goals')
away_goals = df1.groupby('AwayTeam')['FTAG'].sum().reset_index(name='Away Goals')

home_m = df1.groupby('HomeTeam').size().reset_index(name = 'Total Matches')

# Merge data
team_goals = home_goals.merge(away_goals, how='outer', left_on='HomeTeam', right_on='AwayTeam')
team_goals = team_goals.merge(home_m, how='outer', left_on='HomeTeam', right_on='HomeTeam')

# HomeTeam 19 m, AwayTeam 19 m
team_goals['Total Matches'] = team_goals['Total Matches']*2

# Drop column, rename 'HomeTeam', and set 'Team' as the index
team_goals = team_goals.drop(['AwayTeam'], axis=1)
team_goals.rename(columns={"HomeTeam": "Team"}, inplace=True)
team_goals = team_goals.set_index('Team')

# Calculate Average Goals per Game
team_goals["Total Goals"] = team_goals["Home Goals"] + team_goals["Away Goals"]
team_goals["Average Goals per Game"] = team_goals["Total Goals"] / team_goals['Total Matches']

# Display the team that scored the most goals
team_goals.sort_values(by = 'Average Goals per Game', ascending = False).iloc[:1,-1:]
```

## 3. Which team has the most clean sheets in this season?

```python
df_cs = df[['HomeTeam','AwayTeam','FTHG','FTAG']]
```

```python
home_cs = df_cs[df_cs['FTAG'] == 0].groupby('HomeTeam')['FTAG'].size().reset_index(name='clean sheets_home')
away_cs = df_cs[df_cs['FTHG'] == 0].groupby('AwayTeam')['FTHG'].size().reset_index(name='clean sheets_away')

team_cs = away_cs.merge(home_cs, how='outer', left_on='AwayTeam', right_on='HomeTeam').fillna(0)
team_cs = team_cs.drop(['HomeTeam'], axis=1)
team_cs.rename(columns={"AwayTeam": "Team"}, inplace=True)

team_cs["Total Clean Sheets"] = team_cs["clean sheets_home"] + team_cs["clean sheets_away"]

# Display team that had the most clean sheets
team_cs[team_cs["Total Clean Sheets"] == team_cs["Total Clean Sheets"].max()]
```

## 4. Which football teams have the winning percentage above the third quartile?
sorted by winning percentage

```python
df_won = df[['HomeTeam','AwayTeam','FTR']]
```

```python
home_won = df_won[df_won['FTR'] == 'H'].groupby('HomeTeam')['FTR'].size().reset_index()
away_won = df_won[df_won['FTR'] == 'A'].groupby('AwayTeam')['FTR'].size().reset_index()

home_m = df_won.groupby('HomeTeam').size().reset_index(name = 'TotalMatches')
```

```python
team_won = home_won.merge(away_won, how='outer', left_on='HomeTeam', right_on='AwayTeam', suffixes=('_home', '_away'))
team_won = team_won.merge(home_m, how='outer', left_on='HomeTeam', right_on='HomeTeam')

# HomeTeam 19 m, AwayTeam 19 m
team_won['TotalMatches'] = team_won['TotalMatches']*2

# Merge data, Drop 'AwayTeam' column, rename 'HomeTeam', and set 'Team' as the index
team_won.drop(['AwayTeam'], axis=1, inplace=True)
team_won.rename(columns={"HomeTeam": "Team"}, inplace=True)
team_won.set_index('Team', inplace=True)

# Calculate total won matches
team_won["TotalWonMatches"] = team_won["FTR_home"] + team_won["FTR_away"]
team_won['WinningRate'] = (team_won["TotalWonMatches"] / team_won['TotalMatches']) * 100
team_won['WinningRate'] = team_won['WinningRate'].round(2)

# Display the result
team_won[team_won['WinningRate'] > team_won['WinningRate'].quantile(0.75)].sort_values(by = 'WinningRate', ascending = False).iloc[:,-2:]
```

## 5. Which three football teams have the highest conversion rates of shots on target and total shots? 
- Sorted by the conversion rate of shots on target, and then by the conversion rate of total shots
- Considering specifically the team's finishing ability with their conversion rate of shots on target (while conversion rates of total shots  reflect overall goal-scoring efficiency)
- Conversion rates in football represent the efficiency of a team in turning their shots into goals, providing insights into their attacking ability.

```python
df_shot = df[['HomeTeam','AwayTeam','FTHG','FTAG','HS','AS','HST','AST']]
```

```python
home_shots = df_shot.groupby('HomeTeam')[['FTHG','HST','HS']].sum().reset_index()
away_shots = df_shot.groupby('AwayTeam')[['FTAG','AST','AS']].sum().reset_index()

# Merge data, Drop 'AwayTeam' column, rename 'HomeTeam', and set 'Team' as the index
team_shots = home_shots.merge(away_shots, how='outer', left_on = 'HomeTeam', right_on = 'AwayTeam').fillna(0)
team_shots.drop(['AwayTeam'], axis=1, inplace=True)
team_shots.rename(columns={"HomeTeam": "Team"}, inplace=True)
team_shots.set_index('Team', inplace=True)

# Calculate 
team_shots["Total Shots"] = team_shots["HS"] + team_shots["AS"]
team_shots["Shots on target"] = team_shots["HST"] + team_shots["AST"]
team_shots["Goals"] = team_shots["FTHG"] + team_shots["FTAG"]

team_shots['Conversion rate of shots'] = ( team_shots['Goals'] / team_shots['Total Shots'] )*100
team_shots['Conversion rate of shots on target'] = ( team_shots['Goals'] / team_shots['Shots on target'] )*100

team_shots['Conversion rate of shots']  = team_shots['Conversion rate of shots'].round(2)
team_shots['Conversion rate of shots on target'] = team_shots['Conversion rate of shots on target'].round(2)

# Display three football teams that have the highest conversion rates of shots on target
team_shots.sort_values(by = ['Conversion rate of shots on target','Conversion rate of shots'], ascending = False).iloc[:3,6:]
```
